```python
# Import libraries:
import numpy as np
import pandas as pd
import seaborn as sns  # optional, but recommended
import matplotlib.pyplot as plt 
from sklearn import datasets
```

```python
cancer = datasets.load_breast_cancer()
# Check keys:
cancer.keys()
```

```python
# View description:
print(cancer.DESCR)
```

```python
# Check feature names:
cancer.feature_names
```

```python
# And target names:
cancer.target_names
# Note: the target name corresponding to a target (or label) value of i is given by target_names[i].
# Thus we see that 'malignant' is represented by 0 and 'benign' by 1.
```

```python
# Convert data to pandas DataFrame:
df = pd.DataFrame(cancer.data, columns = cancer.feature_names)
# Selecting a slice
df[:100]
```

```python
# Viewing the 100th row:
df[99:100]
```

# a)
### How many rows and columns are there in this data set?

```python
print("(#rows, #columns)")
print(df.shape)
```

### What do the rows and columns represent?
Rows: data points/instances. Each row represents a breast mass in this case.

Columns: features.

# b)
### How many malignant cases are there in total?

```python
# Add 'target' key for convenience 
df['target'] = cancer.target
df['target']
# How many malignant cases do we have?
target = df['target']
q = np.where(target == 0)[0]
print("There are",len(q),"malignant cases.")
print(q)
```

# c)
### Make a scatterplot of ‘mean compactness’ and ‘mean perimeter’. Try to color the points by their label (i.e. malignant or benign). You may find scatter (from matplotlib) or jointplot (from seaborn) useful for this.

```python
# Add 'diagnosis' key for convenience 
df['diagnosis'] = np.array([cancer.target_names[i] for i in cancer.target])

# Plot a scatter plot of 'mean compactness' and 'mean perimeter'
sns.scatterplot(x='mean compactness', y='mean perimeter', hue='diagnosis', data=df)
```

```python
# Alternatively in pure matplotlib:
scatter = plt.scatter(df['mean compactness'], df['mean perimeter'], c=df['target'])
plt.xlabel('mean compactness')
plt.ylabel('mean perimeter')
plt.legend(handles=scatter.legend_elements()[0], labels=list(cancer.target_names))
```

# d)
### Make pairwise scatterplots of ‘mean texture’, ‘mean perimeter’, ‘mean compactness’, and ‘mean radius’. You may find scatter matrix (from matplotlib) or pairplot (from seaborn) useful for this. 

```python
# A pairwise scatter plot of 'mean texture', 'mean perimeter', 'mean compactness', and 'mean radius'
sns.pairplot(df, vars=['mean texture', 'mean perimeter', 'mean compactness', 'mean radius'], hue='diagnosis')
```

#### Do you notice that two of these features seem to be particularly correlated? Why might this be?

"mean perimeter" and "mean radius" are highly correlated. This is because of the mathematical relationship of radius and perimeter. Denoting the perimiter as $p$ and the radius as $r$, $p = 2r\pi$. Likewise, $r=\frac{p}{2\pi}$
